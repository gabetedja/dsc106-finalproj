```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import xarray as xr
import zarr
import gcsfs

xr.set_options(display_style='html')
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
plt.rcParams['figure.figsize'] = 12, 6
```

## Browse Catalog

The data catatalog is stored as a CSV file. Here we read it with Pandas.

```python
df = pd.read_csv('https://storage.googleapis.com/cmip6/cmip6-zarr-consolidated-stores.csv')
df.head()
```

The columns of the dataframe correspond to the CMI6 controlled vocabulary. A beginners' guide to these terms is available in [this document](https://docs.google.com/document/d/1yUx6jr9EdedCOLd--CPdTfGDwEwzPpCF6p1jRmqx-0Q).

Here we filter the data to find monthly surface air temperature for historical experiments.

```python
df_ta = df.query("activity_id=='CMIP' & table_id == 'Amon' & variable_id == 'tas' & experiment_id == 'historical'")
df_ta
```

Now we do further filtering to find just the models from NCAR.

```python
df_ta_ncar = df_ta.query('institution_id == "NCAR"')
df_ta_ncar
```

## Load Data

Now we will load a single store using gcsfs, zarr, and xarray.

```python
# this only needs to be created once
gcs = gcsfs.GCSFileSystem(token='anon')

# get the path to a specific zarr store (the first one from the dataframe above)
zstore = df_ta_ncar.loc[59875].zstore

# create a mutable-mapping-style interface to the store
mapper = gcs.get_mapper(zstore)

# open it using xarray and zarr
ds = xr.open_zarr(mapper, consolidated=True)
ds
```

Plot a map from a specific date.

```python
ds.tas.sel(time='1950-01').squeeze().plot()
```

Create a timeseries of global-average surface air temperature. For this we need the area weighting factor for each gridpoint.

```python
df_area = df.query("variable_id == 'areacella' & source_id == 'CESM2'")
ds_area = xr.open_zarr(gcs.get_mapper(df_area.zstore.values[0]), consolidated=True)
ds_area
```

```python
total_area = ds_area.areacella.sum(dim=['lon', 'lat'])
ta_timeseries = (ds.tas * ds_area.areacella).sum(dim=['lon', 'lat']) / total_area
ta_timeseries
```

By default the data are loaded lazily, as Dask arrays. Here we trigger computation explicitly.

```python
ta_timeseries.load()
```

```python
ta_timeseries.plot(label='monthly')
ta_timeseries.rolling(time=12).mean().plot(label='12 month rolling mean')
plt.legend()
plt.title('Global Mean Surface Air Temperature')
```

Or, we can select e.g. San Diego

```python
sd_time_series = ds.tas.sel(lon=360-117.1611, lat=32.7157, method='nearest')
```

```python
(sd_time_series.rolling(time=12).mean()-sd_time_series.sel(time=slice("1850","1900")).mean()).plot(label='12 month rolling mean')
plt.axhline(0.)
```

```python
# this only needs to be created once
gcs = gcsfs.GCSFileSystem(token='anon')

# get the path to a specific zarr store (the first one from the dataframe above)
zstore = df_ta_2.loc[59875].zstore

# create a mutable-mapping-style interface to the store
mapper = gcs.get_mapper(zstore)

# open it using xarray and zarr
ds = xr.open_zarr(mapper, consolidated=True)
ds
```

```python
df_area = df.query("variable_id == 'areacella' & source_id == 'CESM2'")
ds_area = xr.open_zarr(gcs.get_mapper(df_area.zstore.values[0]), consolidated=True)
ds_area
```

```python
tas = ds['tas'] - 273.15 #deg celsius
tas_2010 = tas.sel(time=tas.time.dt.year == 2010)
```

```python
w = np.cos(np.deg2rad(tas.lat))  
tas_zonal = tas_2010.weighted(w).mean(dim='lon')
df_tas = tas_zonal.groupby('time.month').mean(dim='time').to_dataframe().reset_index()
df_tas.to_csv('tas_vs_lat_2010.csv', index=False)
```

```python
df_psl = df.query("variable_id == 'psl' & source_id == 'CESM2'")
ds_psl = xr.open_zarr(gcs.get_mapper(df_psl.zstore.values[0]), consolidated=True)

ds_psl    # sanity check; should show dims (time, lat, lon) or similar
```

```python
psl_2010 = ds_psl['psl'].sel(time=ds_psl.time.dt.year == 2010)
df_psl_map = psl_2010.to_dataframe(name='psl').reset_index()
df_psl_map.to_csv('psl_lon_lat_2010.csv', index=False)
```

```python
w2 = np.cos(np.deg2rad(psl_2010.lat))  
psl_zonal = psl_2010.weighted(w2).mean(dim='lon')
df_psl = psl_zonal.groupby('time.month').mean(dim='time').to_dataframe().reset_index()
df_psl.to_csv('psl_vs_lat_2010.csv', index=False)
```

```python
df_o3 = df.query("variable_id == 'o3' & source_id == 'CESM2'")
ds_o3 = xr.open_zarr(gcs.get_mapper(df_o3.zstore.values[0]), consolidated=True)

o3_2010 = (
    ds_o3['o3']
    .sel(time=ds_o3.time.dt.year == 2010, plev=ds_o3.plev[0])   # or .mean('lev')
)
df_o3_map = o3_2010.to_dataframe(name='o3').reset_index()
```

```python
df_o3_map.to_csv('o3_plev_2010.csv', index=False)
```

```python
w3 = np.cos(np.deg2rad(o3_2010.lat))  
o3_zonal = o3_2010.weighted(w3).mean(dim='lon')
df_o3 = o3_zonal.groupby('time.month').mean(dim='time').to_dataframe().reset_index()
df_o3.to_csv('o3_vs_lat_2010.csv', index=False)
```

```python
df_dust_like = df[df['variable_id'].str.contains('dust', case=False, na=False)]
df_dust_like[['variable_id', 'source_id']].drop_duplicates()
df_dust_like.head()
```

```python
df_dust = df.query("variable_id == 'dustvar' & source_id == 'CESM2'")
ds_dust = xr.open_zarr(gcs.get_mapper(df_dust.zstore.values[0]), consolidated=True)

dust_2010 = ds_dust['dustvar'].sel(time=ds_dust.time.dt.year == 2010).mean('time')
df_dust_map = dust_2010.to_dataframe(name='dust').reset_index()
df_dust_map.to_csv('dust_lon_lat_2010.csv', index=False)
```
